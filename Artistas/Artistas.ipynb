### Centralidade
Existem diversas medidas de centralidades que podem ser tiradas de um grafo. Para Medir as centralidades em diferentes métodos e apresentar o resultado, foi feita uma função para usar os métodos como variáveis e retornar uma lista com as medidas para serem colocadas em um gráfico. A função é chamada de *centralidade()*, com assinatura *centralidade(metodo,genero)*.

```python
resultado_grau = centralidade()
resultado_in = centralidade(nx.in_degree_centrality)
resultado_out = centralidade(nx.out_degree_centrality)
resultado_between = centralidade(nx.betweenness_centrality)
```

```python
resultados = [resultado_grau,resultado_in,resultado_out,resultado_between]
for resultado in resultados:
    lista_histograma(list(resultado.values()),title='Resultados Centralidade',xlabel='Resultado')
```

```python
resultados = [resultado_grau,resultado_in,resultado_out,resultado_between]
for resultado in resultados:
    resultado_disrupcao = [disruption_genre.loc[r].get('disruption')
                      if r in disruption_genre.index 
                      else None
                      for r in resultado]
    
    resultado_valor = [ resultado[r]
                      if r in disruption_genre.index 
                      else None
                      for r in resultado.keys()]
    
    listas_dispersao(resultado_disrupcao,resultado_valor,xlabel='Disrupção',ylabel='Valor Centralidade')
```
